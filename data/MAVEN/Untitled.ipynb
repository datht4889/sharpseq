```python
import json
from collections import Counter
from matplotlib import pyplot as plt
import numpy as np
```

```python
with open("dev.origin") as f:
    data = [json.loads(line) for line in f]
```

```python
a = Counter([i["label"] for i in data]).most_common()
a
```

```python
keys = [i[0] for i in a]
values = [i[1] for i in a]
```

```python
(np.array(values)/sum(values)*100).tolist()
```

```python
plt.figure(figsize=(15,15))
plt.pie(values, labels=keys)
```
