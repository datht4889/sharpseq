```python
llama2 = AutoTokenizer.from_pretrained("McGill-NLP/LLM2Vec-Meta-Llama-3-8B-Instruct-mntp")
```

```python
def read_data(path):
    with open(path) as f:
        data = []
        for line in f:
            data.append(json.loads(line))
        return data

def process(item):
    # import pdb
    # pdb.set_trace()
    input_ids = item["piece_ids"]
    text = tokenizer.decode(input_ids, skip_special_tokens=True).strip()
    first = tokenizer.decode(input_ids[:item["span"][0]], skip_special_tokens=True).strip()
    second = tokenizer.decode(input_ids[:item["span"][1]+1], skip_special_tokens=True).strip()
    input_ids = llama2(text)["input_ids"]
    first = llama2(first)["input_ids"]
    second = llama2(second)["input_ids"]
    span = [len(first),  len(second)-1]
    if span[0] >= len(input_ids):
        print("out of indices")
        span[0] = len(input_ids) - 1

    item["piece_ids"] = input_ids
    item["span"] = span

def convert(path, output):
    data = read_data(path)
    #new_data = []
    for i in data:
        process(i)
        # import pdb
        # pdb.set_trace()
    # import pdb
    # pdb.set_trace()
    with open(output, "w") as f:
        for i in data:
            f.write(json.dumps(i)+"\n")
```
